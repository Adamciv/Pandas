# ffil - bfill - interpolate( )

1-  ffil:

**ffil (forward fill) is a method that is used to fill missing values in a Pandas Series or
DataFrame with the previous non-null value.**

2- bfil :

**In Pandas, you can use the bfill (backward fill) method to fill missing values in a Pandas Series or
DataFrame with the next non-null value.**



```python
# ffil
```

```python
df3=df.copy()
```

```python
df3['Age'].fillna(method='ffill',inplace=True)
```

```python
df3
```

```python
# bfill
```

```python
df3['Height'].bfill(inplace=True)
df3
```
